```python
import keras
(train_input, train_target), (test_input, test_target)=\
  keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

train_scaled = train_input/255.0
train_scaled = train_scaled.reshape(-1,28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42
)
```

```python
inputs = keras.layers.Input(shape=(784,))
dense1=keras.layers.Dense(100, activation='sigmoid')
dense2=keras.layers.Dense(10,activation='softmax')
```

```python
model=keras.Sequential([inputs,dense1,dense2])
```

```python
model.summary()
```

시그모이드(sigmoid)와 소프트맥스(softmax)는 뉴런의 출력을 변형해 주는 활성화 함수(activation function) 입니다.​
신경망이 단순한 선형함수에 그치지 않고, “확률처럼 해석 가능한 값”을 내도록 만드는 핵심 역할을 합니다.​

시그모이드
출력: 0~1 사이의 단일 값.​
각 출력이 서로 독립적이라, 여러 개의 시그모이드 뉴런을 쓰면 “여러 개에 동시에 1이 나와도 상관없는” 멀티 라벨 구조에 적합합니다.​

소프트맥스
출력: 여러 값이지만 합이 1인 확률 분포.​
한 샘플이 여러 클래스 중 정확히 하나만 속한다고 가정할 때(숫자 0~9, 옷 종류 10개 등) 적합합니다.​

```python
model=keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(100, activation='sigmoid'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```

옵티마이저?
손실 함수(loss)를 줄이기 위해, 가중치(파라미터)를 어떻게 업데이트할지 결정하는 알고리즘입니다.
같은 모델 구조라도 “어떤 옵티마이저를 쓰느냐”에 따라 학습 속도, 수렴 안정성, 최종 성능이 꽤 달라집니다.

1. 왜 옵티마이저가 필요한가?
신경망의 목표는 손실 함수
L(w)가 최소가 되도록 가중치 w를 찾는 것입니다.
2. 가장 기본: 경사 하강법과 SGD 계열
경사 하강법(Gradient Descent)

모든 데이터에 대해 손실을 계산하고 기울기를 구한 뒤, 그 기울기의 반대 방향으로 한 번 업데이트합니다.

정확하지만, 데이터가 클수록 한 스텝이 너무 느려지고 메모리를 많이 씁니다.

SGD(Stochastic Gradient Descent)

전체 데이터 대신, 미니배치(일부 샘플) 로 기울기를 계산해서 훨씬 빠르게 자주 업데이트합니다.

계산량이 적고 온라인 학습에도 쓰기 좋지만, 기울기가 들쭉날쭉해서 경로가 많이 흔들릴 수 있습니다.

대부분의 “현대 옵티마이저”들은 SGD를 기반으로, 더 잘/더 안정적으로 내려가도록 개선한 버전이라고 생각하면 됩니다.

3. 대표적인 옵티마이저들 직관
각 옵티마이저는 “산을 내려갈 때 어떻게 걸을지”에 대한 전략 차이로 이해하면 편합니다.

SGD (또는 vanilla SGD)

형태로, 현재 미니배치에서 얻은 기울기만 보고 한 스텝 이동합니다.

장점: 단순, 구현 쉬움.

단점: 경로가 요동치고, 학습률(learning rate)에 매우 민감합니다.

Momentum

이전에 움직인 속도를 기억해서 관성처럼 더해 줍니다.

비유: 공을 언덕에 굴리면, 기울기가 약간 바뀌어도 관성 때문에 같은 방향으로 계속 굴러가는 것과 비슷합니다.

효과: local minimum/안장점에서 잘 빠져나오고, 지그재그가 줄어들어 더 빠르게 수렴할 수 있습니다.

RMSProp

파라미터마다 “최근에 기울기 제곱이 얼마나 컸는지”를 보고, 각 파라미터별로 학습률을 다르게 조절합니다.

자주 크게 변하는 파라미터는 학습률을 줄이고, 잘 안 변하던 파라미터는 학습률을 키우는 식입니다.

비유: 급경사에서는 보폭을 줄이고, 평평한 곳에서는 보폭을 크게 하는 느낌.

Adam (Adaptive Moment Estimation)

Momentum + RMSProp 를 섞어 놓은, 요즘 가장 많이 쓰이는 옵티마이저입니다.

기울기의 평균(1차 모멘트)과 제곱의 평균(2차 모멘트)을 모두 추적해서, 방향과 보폭을 동시에 “똑똑하게” 결정합니다.

장점: 보통 튜닝 없이도 잘 동작하고, 다양한 모델에서 안정적으로 빠르게 수렴하는 편이라 기본값처럼 많이 사용됩니다.

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28,28)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(100,activation='relu'))
model.add(keras.layers.Dense(10,activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```

```python
model.evaluate(val_scaled, val_target)
```
